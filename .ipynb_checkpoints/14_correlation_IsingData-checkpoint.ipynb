```python
# Top Eignvectors of the C_ij matrix:
def topEV(data,method='Covariance'):
    if method is 'Covariance':
        #Find the covariance matrix:
        covMAT = np.cov(data)
    elif method is 'Correlation':
        covMAT = np.corrcoef(data)
    Eval,Evec = np.linalg.eig(covMAT)
    return Evec[:,np.argmax(Eval)],np.max(Eval)
```

```python
XBM      = []
Xdata    = []
Xresults = []
XRBM     = []
mat_topEV = np.zeros([len(beta),int(N[0])])
mat_topEV_corr = np.zeros([len(beta),int(N[0])])
mat_topEval = np.zeros(len(beta))
mat_topEVal_corr = np.zeros(len(beta))

for i in np.arange(len(beta)):
    #Generate spins using Gibbs Sampling MCMC
    XBM.append(bm.BM(N=N[i],nature='Spin'))

    #Define Couplings
    XBM[i].layer.couplings =  beta[i]*(np.eye(N=N[i],k=1) + np.eye(N=N[i],k=-1) )
    XBM[i].layer.couplings[0,-1] += beta[i]
    XBM[i].layer.couplings[-1,0] += beta[i]

    #Generating the ising model data
    Xdata.append(XBM[i].gen_data(Nchains=1,Nthermalize=0,Lchains=Num_configs,Nstep=Nstep[i])) # N steps.
    
    #Calculate Top eignevectors and store it in a matrix
    mat_topEV[i,:],mat_topEval[i] = topEV(Xdata[i].T)
    mat_topEV_corr[i,:],mat_topEVal_corr[i] = topEV(Xdata[i].T,method='Correlation')
```

```python
XBM7      = []
Xdata7    = []
Xresults7 = []
XRBM7     = []
mat_topEV7 = np.zeros([len(beta7),int(N7[0])])
mat_topEV7_corr7 = np.zeros([len(beta7),int(N7[0])])
mat_topEval7 = np.zeros(len(beta7))
mat_topEval7_corr7 = np.zeros(len(beta7))


for i in np.arange(len(beta7)):
    #Generate spins using Gibbs Sampling MCMC
    XBM7.append(bm.BM(N=N7[i],nature='Spin'))

    #Define Couplings
    XBM7[i].layer.couplings =  beta7[i]*(np.eye(N=N7[i],k=1) + np.eye(N=N7[i],k=-1) )
    XBM7[i].layer.couplings[0,-1] += beta7[i]
    XBM7[i].layer.couplings[-1,0] += beta7[i]

    #Generating the ising model data
    Xdata7.append(XBM7[i].gen_data(Nchains=1,Nthermalize=0,Lchains=int(Num_configs7[i]),Nstep=Nstep7[i])) # N steps.
    
    #Calculate Top eignevectors and store it in a matrix
    mat_topEV7[i,:],mat_topEval7[i] = topEV(Xdata7[i].T)
    mat_topEV7_corr7[i,:],mat_topEval7_corr7[i] = topEV(Xdata7[i].T,method='Correlation')
```

```python
plt.figure(figsize=(10,5))
plt.scatter(float(N7[0])/Num_configs7.astype('float'),mat_topEval7_corr7)
plt.plot(float(N7[0])/Num_configs7.astype('float'),(1+(float(N7[0])/Num_configs7.astype('float'))**0.5)**2)
plt.plot(float(N7[0])/Num_configs7.astype('float'),np.zeros(len(Num_configs7)) + (1+np.tanh(0.1))/(1-np.tanh(0.1)))
plt.xscale('log')
#plt.yscale('log')
plt.xlabel('Num_visible_units/Num_Samples')
plt.ylabel('Top Eignevalue of Correlation Matrix')
plt.title('beta = 0.1')
```
