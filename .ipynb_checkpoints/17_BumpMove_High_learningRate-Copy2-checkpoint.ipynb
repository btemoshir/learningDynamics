```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

## Make the landscape extremely smooth by increasing the number of samples and keep the same learning rate (only one hidden unit first): 

```python
%%capture capt4
#Supress the output of the cell and save it to capt1

N = 100
beta = 1.
Num_configs4 = 1000000

#Generate spins using Gibbs Sampling MCMC
BM4 = bm.BM(N=N,nature='Spin')

#Define Couplings
BM4.layer.couplings =  beta* (np.eye(N=N,k=1) + np.eye(N=N,k=-1) )
BM4.layer.couplings[0,-1] += beta
BM4.layer.couplings[-1,0] += beta

data4 = BM4.gen_data(Nchains=1,Nthermalize=0,Lchains=Num_configs4,Nstep=400) # N steps.

#Train the model
RBM4 = rbm.RBM(visible = 'Spin', n_v=N, n_h = 1, hidden = 'Spin',zero_field = True)
results4 = RBM4.fit(data4,n_iter=250,verbose=0,learning_rate = 0.1,record=['W'],record_interval=1000,batch_size=100,N_MC= 20, lr_decay=False,batch_norm=False);
```

```python
plt.plot(np.array(results4['W'])[-200,0,:])
```

```python
peakCentre4 = np.zeros([np.shape(np.array(results4['W']))[0],np.shape(np.array(results4['W']))[1]])
peakCentre4 = np.argmax(abs(np.array(results4['W'])),axis=2)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(np.arange(2500),peakCentre4,marker='.')
```

```python
PBC_corc = np.copy(peakCentre4[1400:-1,0])
PBC_corc[PBC_corc<50] = PBC_corc[PBC_corc<50] + 100
PBC_corc = np.append(peakCentre4[0:1401,0],PBC_corc)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(np.arange(2500),PBC_corc,marker='.')
```

#### Replace the ensemble average for MSD by a time aveareg for this trajectory:

```python
MSD = np.zeros(len(peakCentre4)/2)
for i in np.arange(len(peakCentre4)/2):
    MSD[i] = (np.sum((PBC_corc - np.roll(PBC_corc,i))**2))/(len(PBC_corc))
linear_coeff = np.polyfit(np.arange(2500/2),MSD,1)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(np.arange(2500/2),MSD,marker='o',s=1,label='MSD from trajectory')
plt.plot(np.arange(2500/2),np.arange(2500/2)*linear_coeff[0]+linear_coeff[1],'r',label='Linear Fit')
plt.legend()
plt.xlabel('time (1/10th of an epoch)')
plt.ylabel('Average Means Square Displacement (A.U.)')
```
