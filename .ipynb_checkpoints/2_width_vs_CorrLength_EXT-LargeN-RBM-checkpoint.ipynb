```python
def beta_from_corrLength(corrLen):
    beta = np.tan((np.exp(-1./corrLen)))
    return beta
```

```python
beta = 1.07
```

```python
-1./np.log(np.arctan(beta))
```

```python
Num_configs = 25000
Nstep       = int(beta*1000/1.5)
N           = np.arange(100,1100,100)
N           = N.astype(int)
N
print('Num_steps = ' + np.str(Nstep))
```

## True Error bars with the same data but different fitting

```python
#%%capture captX

Xresults_big = []

XBM      = []
Xdata    = []
Xresults = []
XRBM     = []

Num_reps = 10

for i in range(len(N)):
    
    Xresults = []
    XRBM     = []
    
    #Generate spins using Gibbs Sampling MCMC
    XBM.append(bm.BM(N=N[i],nature='Spin'))

    #Define Couplings
    XBM[i].layer.couplings =  beta*(np.eye(N=N[i],k=1) + np.eye(N=N[i],k=-1) )
    XBM[i].layer.couplings[0,-1] += beta
    XBM[i].layer.couplings[-1,0] += beta

    #Generating the ising model data
    Xdata.append(XBM[i].gen_data(Nchains=1,Nthermalize=0,Lchains=Num_configs,Nstep=Nstep)) # N steps.

    
    for j in np.arange(Num_reps):
        
        #Train the model
        XRBM.append(rbm.RBM(visible = 'Spin', n_v=N[i], n_h = 1, hidden = 'Spin',zero_field = True))
        Xresults.append(XRBM[j].fit(Xdata[i],n_iter=250,verbose=0,learning_rate = 0.001,record=['W'],record_interval=1000,batch_size=10,N_MC= 20, lr_decay=False,batch_norm=False))
        
    Xresults_big.append(Xresults)
```

```python
import peakWidth
XFWHM_avg = []
XFWHM_error = []

for i in np.arange(len(N)):
    
    t1 = np.zeros(Num_reps)
    
    for j in range(Num_reps):
        t1[j] = np.array(peakWidth.peakWidth(weight=np.array(Xresults_big[j][i]['W'])[-1,0,:],plot=True))
        
    XFWHM_avg.append(np.mean(t1))
    XFWHM_error.append(np.std(t1))
```

```python
#Plot FWHM as a function of the correlation length
import numpy as np

plt.figure(figsize=(10,5))
#Xfit = np.polyfit(corrLengths,XFWHM_avg, deg=1)
#plt.plot(corrLengths, Xfit[0] * corrLengths + Xfit[1], color='red')
plt.errorbar(N,XFWHM_avg,yerr=XFWHM_error,xlolims=0,fmt='o',ecolor='g')
#plt.plot(N,XFWHM_avg,'o')
#plt.yscale('log')
#plt.xscale('log')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Number of visible units(N)',fontsize=14)
plt.ylabel('Width of the peak (FWHM)',fontsize=14)
plt.show()
```

```python
#Plot FWHM as a function of the correlation length
import numpy as np

plt.figure(figsize=(10,5))
Xfit = np.polyfit(np.log(N),np.log(XFWHM_avg/N), deg=1)
plt.plot(N, np.exp(Xfit[0]*np.log(N) + Xfit[1]), color='red')
plt.text(300,0.1,'~ '+(np.array2string(np.round(Xfit[0],3))),fontsize=14,color='r')
#plt.errorbar(N,XFWHM_avg/N,yerr=XFWHM_error,xlolims=0,fmt='o',ecolor='g')
plt.plot(N,XFWHM_avg/N,'o')
plt.ylim((10**-2,0.2))
plt.yscale('log')
plt.xscale('log')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Number of visible units(N)',fontsize=14)
plt.ylabel('Width of the peak (FWHM)/N',fontsize=14)
plt.show()
```
